```python
def explore_folder(path, indent=0):
    """Recursively explore folder contents"""
    items = os.listdir(path)
    for item in items:
        full_path = os.path.join(path, item)
        if os.path.isdir(full_path):
            print('  ' * indent + f'ðŸ“ {item}/')
            explore_folder(full_path, indent + 1)
        else:
            print('  ' * indent + f'ðŸ“„ {item}')

print(f"Exploring folder: {folder_path}")
explore_folder(folder_path)
```

```python
def peek_at_file(file_path, max_lines=5):
    """Display the first few lines of a file"""
    print(f"\nPeeking at: {file_path}")
    try:
        if file_path.lower().endswith('.csv'):
            df = pd.read_csv(file_path, nrows=max_lines)
            display(df)
        elif file_path.lower().endswith(('.xls', '.xlsx')):
            df = pd.read_excel(file_path, nrows=max_lines)
            display(df)
        else:
            with open(file_path, 'r') as f:
                for i, line in enumerate(f):
                    if i >= max_lines:
                        break
                    print(line.strip())
    except Exception as e:
        print(f"Error reading file: {e}")

# Peek at the first 3 data files
for file in data_files[:6]:
    peek_at_file(file)
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# 1. Load and merge data (as previously done)
def load_data():
    patient_info = pd.read_csv('/content/drive/MyDrive/hyperaktiv_with_controls/patient_info.csv', sep=';')
    features = pd.read_csv('/content/drive/MyDrive/hyperaktiv_with_controls/features.csv', sep=';')
    cpt_data = pd.read_csv('/content/drive/MyDrive/hyperaktiv_with_controls/CPT_II_ConnersContinuousPerformanceTest.csv', sep=';')

    # Create base dataset
    adhd_data = patient_info[['ID', 'SEX', 'AGE', 'ADHD', 'ADD', 'CPT_II']].copy()

    # Add CPT scores
    cpt_features = cpt_data[['ID', 'Adhd TScore Omissions', 'Adhd TScore Commissions',
                            'Adhd TScore HitRT', 'Adhd Confidence Index']]
    adhd_data = adhd_data.merge(cpt_features, on='ID', how='left')

    # Add features (only for patients with IDs 1-116)
    features['ID'] = features.index + 1
    adhd_data = adhd_data.merge(features, on='ID', how='inner')  # Inner join to keep only matches

    return adhd_data

# 2. Preprocessing
def preprocess_data(df):
    # Handle missing CPT data (median imputation)
    cpt_cols = ['Adhd TScore Omissions', 'Adhd TScore Commissions',
                'Adhd TScore HitRT', 'Adhd Confidence Index']
    df[cpt_cols] = SimpleImputer(strategy='median').fit_transform(df[cpt_cols])

    # Convert binary variables
    df['SEX'] = df['SEX'].astype(int)  # Ensure proper encoding

    # Feature scaling
    numeric_cols = ['AGE'] + cpt_cols + [col for col in df.columns if col.startswith('feature_')]
    df[numeric_cols] = StandardScaler().fit_transform(df[numeric_cols])

    return df

# 3. Create modeling datasets
def create_model_datasets(target='ADHD', test_size=0.2, random_state=42):
    df = load_data()
    df = preprocess_data(df)

    # Define features and target
    X = df.drop(columns=['ID', 'ADHD', 'ADD', 'CPT_II'])
    y = df[target]

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state, stratify=y)

    return X_train, X_test, y_train, y_test, df.columns.tolist()
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Reshape features into 3D format (samples, timesteps, features)
def reshape_for_cnn(X, timesteps=10):
    """Convert tabular data to pseudo-time-series"""
    n_samples, n_features = X.shape
    n_sequences = n_samples // timesteps
    remainder = n_samples % timesteps

    # Pad if needed
    if remainder != 0:
        pad_size = timesteps - remainder
        X = np.concatenate([X, np.zeros((pad_size, n_features))])

    return X.reshape(-1, timesteps, n_features)

# Example transformation
X_train_cnn = reshape_for_cnn(X_train_full.values, timesteps=10)
X_test_cnn = reshape_for_cnn(X_test_full.values, timesteps=10)
print(f"New CNN input shape: {X_train_cnn.shape}")
```

```python
def create_overlapping_windows(X, y, window=10, stride=5):
    windows = []
    labels = []
    for start in range(0, len(X) - window + 1, stride):
        window_labels = y[start:start+window]
        if 1 in window_labels:  # Keep windows with â‰¥1 ADHD case
            windows.append(X[start:start+window])
            labels.append(1)  # Treat as ADHD-positive window
        else:
            windows.append(X[start:start+window])
            labels.append(0)  # Non-ADHD window
    return np.array(windows), np.array(labels)
```

```python
def create_balanced_windows(X, y, window=10, stride=5):
    windows = []
    labels = []
    for start in range(0, len(X) - window + 1, stride):
        window_labels = y[start:start+window]
        if 0 in window_labels or 1 in window_labels:  # Keep if either class exists
            windows.append(X[start:start+window])
            labels.append(1 if 1 in window_labels else 0)  # Label by majority
    return np.array(windows), np.array(labels)

# Regenerate windows
X_train_cnn, y_train_seq = create_balanced_windows(X_train_full.values, y_train_full.values)
X_val_cnn, y_val_seq = create_balanced_windows(X_val_full.values, y_val_full.values)  # If you have a separate val set
```
