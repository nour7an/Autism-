```python
# ==============================================
# IMPORT LIBRARIES
# ==============================================
import pandas as pd
import numpy as np
import os
import re
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, GlobalAveragePooling1D
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.initializers import he_normal

# ==============================================
# DATA LOADING AND PREPROCESSING
# ==============================================
def preprocess_patient_data(df):
    # Convert comma decimals to points
    decimal_cols = ['ACC_DAYS', 'HRV_HOURS', 'WURS', 'ASRS', 'MADRS', 'HADS_A', 'HADS_D']
    for col in decimal_cols:
        if col in df.columns:
            df[col] = df[col].astype(str).str.replace(',', '.').astype(float)

    # Convert binary columns
    binary_cols = ['ADHD', 'ADD', 'BIPOLAR', 'UNIPOLAR', 'ANXIETY',
                   'SUBSTANCE', 'OTHER', 'CT', 'MDQ_POS', 'MED']
    for col in binary_cols:
        if col in df.columns:
            df[col] = df[col].fillna(0).astype(int)

    # Fill numeric columns with median
    numeric_cols = ['AGE', 'ACC_DAYS', 'HRV_HOURS', 'WURS', 'ASRS',
                    'MADRS', 'HADS_A', 'HADS_D']
    for col in numeric_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
            df[col] = df[col].fillna(df[col].median())

    # Convert time columns to total seconds
    time_cols = ['ACC_TIME', 'HRV_TIME']
    for col in time_cols:
        if col in df.columns:
            df[col] = df[col].astype(str)
            df[col] = df[col].replace('nan', np.nan)
            df[col] = pd.to_datetime(df[col], format='%H:%M:%S', errors='coerce')
            df[col+'_SEC'] = df[col].dt.hour * 3600 + df[col].dt.minute * 60 + df[col].dt.second
            df.drop(col, axis=1, inplace=True)

    # Drop columns with too many missing values
    cols_to_drop = ['MED_Antidepr', 'MED_Moodstab', 'MED_Antipsych',
                   'MED_Anxiety_Benzo', 'MED_Sleep', 'MED_Analgesics_Opioids',
                   'MED_Stimulants']
    for col in cols_to_drop:
        if col in df.columns:
            df.drop(col, axis=1, inplace=True)

    return df

# ==============================================
# ACTIVITY DATA PROCESSING
# ==============================================
def load_activity_data(patient_ids, base_path):
    activity_data = []
    valid_ids = []

    for pid in patient_ids:
        file_path = f"{base_path}/patient_activity_{pid}.csv"
        try:
            # Try loading with different formats
            try:
                df = pd.read_csv(file_path, sep=';')
            except:
                df = pd.read_csv(file_path)

            # Handle cases where header might be in first row
            if len(df.columns) == 1 and ';' in df.columns[0]:
                df = pd.read_csv(file_path, sep=';', header=None)
                df.columns = ['TIMESTAMP', 'ACTIVITY']

            # Ensure we have required columns
            if 'ACTIVITY' not in df.columns and len(df.columns) >= 2:
                df.columns = ['TIMESTAMP', 'ACTIVITY'] + list(df.columns[2:])

            if 'ACTIVITY' in df.columns:
                # Process activity values
                activity = pd.to_numeric(df['ACTIVITY'], errors='coerce').dropna().values

                if len(activity) > 10:  # Minimum samples
                    # Robust scaling
                    q75, q25 = np.percentile(activity, [75, 25])
                    scaled = (activity - np.median(activity)) / (q75 - q25 + 1e-8)

                    # Pad/truncate to fixed length
                    fixed_length = 1000
                    if len(scaled) > fixed_length:
                        processed = scaled[:fixed_length]
                    else:
                        processed = np.pad(scaled, (0, fixed_length - len(scaled)),
                                         mode='constant', constant_values=0)

                    # 3-channel format
                    activity_data.append(np.stack([processed]*3, axis=-1))
                    valid_ids.append(pid)
                    print(f"✓ Processed patient {pid}")
                else:
                    print(f"× Insufficient data for patient {pid}")
            else:
                print(f"× Missing ACTIVITY column for patient {pid}")
        except Exception as e:
            print(f"! Error processing patient {pid}: {str(e)}")

    if not activity_data:
        print("\nCRITICAL: No activity data processed successfully")
        return np.array([]), []

    return np.array(activity_data), valid_ids

# ==============================================
# FOCAL LOSS IMPLEMENTATION
# ==============================================
def focal_loss(alpha=0.25, gamma=2.0):
    def loss_fn(y_true, y_pred):
        pt = tf.where(tf.equal(y_true, 1), y_pred, 1-y_pred)
        return -tf.reduce_mean(alpha * tf.pow(1. - pt, gamma) * tf.math.log(pt + 1e-8))
    return loss_fn

# ==============================================
# MAIN EXECUTION
# ==============================================
if __name__ == "__main__":
    # Load and preprocess patient info
    patient_info = pd.read_csv(
        '/content/drive/MyDrive/hyperaktiv_with_controls/patient_info.csv',
        delimiter=';',
        engine='python'
    )
    patient_info_clean = preprocess_patient_data(patient_info)

    # Load activity data
    activity_data_path = '/content/drive/MyDrive/hyperaktiv_with_controls/activity_data'
    X_activity, valid_ids = load_activity_data(patient_info_clean['ID'], activity_data_path)

    if len(X_activity) == 0:
        raise ValueError("No valid activity data loaded - cannot proceed")

    # Filter patient info to only those with valid activity data
    patient_info_filtered = patient_info_clean[patient_info_clean['ID'].isin(valid_ids)].copy()
    y = patient_info_filtered['ADHD'].values

    # Prepare tabular features
    tabular_features = ['SEX', 'AGE', 'ASRS', 'WURS', 'HADS_A', 'HADS_D',
                       'ACC', 'ACC_DAYS', 'HRV', 'HRV_HOURS', 'CPT_II',
                       'ADD', 'BIPOLAR', 'UNIPOLAR', 'ANXIETY', 'CT',
                       'ACC_TIME_SEC', 'HRV_TIME_SEC']
    tabular_features = [col for col in tabular_features if col in patient_info_filtered.columns]
    X_tabular = patient_info_filtered[tabular_features].copy()

    # Scale numeric features
    numeric_cols = ['AGE', 'ASRS', 'WURS', 'HADS_A', 'HADS_D',
                    'ACC_DAYS', 'HRV_HOURS', 'ACC_TIME_SEC', 'HRV_TIME_SEC']
    numeric_cols = [col for col in numeric_cols if col in X_tabular.columns]

    if numeric_cols:
        X_tabular[numeric_cols] = X_tabular[numeric_cols].astype(float)
        scaler = RobustScaler()
        X_tabular.loc[:, numeric_cols] = scaler.fit_transform(X_tabular[numeric_cols])

    X_tabular = X_tabular.values

    # Train-test split
    X_activity_train, X_activity_test, X_tabular_train, X_tabular_test, y_train, y_test = train_test_split(
        X_activity, X_tabular, y, test_size=0.2, random_state=42, stratify=y)

    # Calculate class weights
    class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
    class_weights = dict(enumerate(class_weights))

    # Build model
    activity_input = Input(shape=(1000, 3))
    x = Conv1D(32, 10, activation='relu', kernel_initializer=he_normal())(activity_input)
    x = MaxPooling1D(2)(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)

    x = Conv1D(64, 5, activation='relu', kernel_initializer=he_normal())(x)
    x = GlobalAveragePooling1D()(x)
    activity_output = Dense(16, activation='relu')(x)

    tabular_input = Input(shape=(X_tabular.shape[1],))
    tabular_output = Dense(16, activation='relu', kernel_initializer=he_normal())(tabular_input)
    tabular_output = BatchNormalization()(tabular_output)

    combined = concatenate([activity_output, tabular_output])
    z = Dense(32, activation='relu', kernel_initializer=he_normal())(combined)
    z = Dropout(0.4)(z)
    output = Dense(1, activation='sigmoid')(z)

    model = Model(inputs=[activity_input, tabular_input], outputs=output)

    # Compile model
    model.compile(
        optimizer=Adam(learning_rate=0.0001, clipnorm=1.0),
        loss=focal_loss(),
        metrics=['accuracy',
                 tf.keras.metrics.AUC(name='auc'),
                 tf.keras.metrics.Precision(name='precision'),
                 tf.keras.metrics.Recall(name='recall')]
    )

    # Callbacks
    callbacks = [
        EarlyStopping(patience=10, restore_best_weights=True, monitor='val_auc', mode='max'),
        ReduceLROnPlateau(factor=0.5, patience=5, min_lr=1e-6),
        ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_auc')
    ]

    # Train model
    history = model.fit(
        [X_activity_train, X_tabular_train],
        y_train,
        validation_data=([X_activity_test, X_tabular_test], y_test),
        epochs=100,
        batch_size=32,
        class_weight=class_weights,
        callbacks=callbacks
    )

    # Replace the model saving and loading code with:

# In the callbacks list:
callbacks = [
    EarlyStopping(patience=10, restore_best_weights=True, monitor='val_auc', mode='max'),
    ReduceLROnPlateau(factor=0.5, patience=5, min_lr=1e-6),
    # Replace this:
ModelCheckpoint('best_model.keras', save_best_only=True, monitor='val_auc', save_format='keras')

# With this (remove save_format):
ModelCheckpoint('best_model.keras', save_best_only=True, monitor='val_auc')
]

# ... (keep the rest of the training code the same) ...

# Change the model loading to:
model = load_model('best_model.keras', custom_objects={'focal_loss': focal_loss})
        # Evaluate
    # ... (other code, no indentation) ...

# THIS LINE SHOULD HAVE ZERO INDENTATION
if __name__ == "__main__":
    # Everything inside indented by 4 spaces
    patient_info = pd.read_csv(...)
    patient_info_clean = preprocess_patient_data(patient_info)

    # ... rest of your main code ...

    # Evaluate model (NO EXTRA INDENTATION)
    y_pred = model.predict([X_activity_test, X_tabular_test])  # <-- Align with the left margin
    y_pred_class = (y_pred > 0.5).astype(int)

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred_class, zero_division=0))

    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_pred_class))
    # Plot ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_pred)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()

    # Plot training history
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['auc'], label='Train AUC')
    plt.plot(history.history['val_auc'], label='Validation AUC')
    plt.title('Model AUC')
    plt.ylabel('AUC')
    plt.xlabel('Epoch')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()
    plt.show()
```

```python
# ==============================================
# STEP 1: Improved Activity Data Loading
# ==============================================
# Update activity data loading to ensure better normalization
def load_activity_data(patient_ids, base_path):
    activity_data = []
    valid_ids = []

    for pid in patient_ids:
        file_path = f"{base_path}/patient_activity_{pid}.csv"
        try:
            df = pd.read_csv(file_path, sep=';')

            # Handle cases where header might be in first row
            if len(df.columns) == 1 and ';' in df.columns[0]:
                df = pd.read_csv(file_path, sep=';', header=None)
                df.columns = ['TIMESTAMP', 'ACTIVITY']

            # Ensure numeric activity values
            activity_values = pd.to_numeric(df['ACTIVITY'], errors='coerce').dropna()

            if len(activity_values) > 10:  # Minimum samples
                # Robust scaling (avoid division by zero)
                scaled = (activity_values - np.median(activity_values)) / \
                        (np.percentile(activity_values, 75) - np.percentile(activity_values, 25) + 1e-8)

                # Pad/truncate to fixed length
                fixed_length = 1000
                if len(scaled) > fixed_length:
                    processed = scaled[:fixed_length]
                else:
                    processed = np.pad(scaled, (0, fixed_length - len(scaled)),
                                     mode='constant', constant_values=0)

                # 3-channel format
                activity_data.append(np.stack([processed]*3, axis=-1))
                valid_ids.append(pid)

        except Exception as e:
            continue

    return np.array(activity_data), valid_ids

# ==============================================
# STEP 2: Load and Verify Data
# ==============================================
# Load activity data
X_activity, valid_ids = load_activity_data(patient_info_clean['ID'], activity_data_path)

if len(X_activity) == 0:
    raise ValueError("Stopping: No valid activity data loaded")

# Filter patient info
patient_info_filtered = patient_info_clean[patient_info_clean['ID'].isin(valid_ids)].copy()
y = patient_info_filtered['ADHD'].values

# ==============================================
# STEP 3: Prepare Tabular Features (Fixed)
# ==============================================
tabular_features = ['SEX', 'AGE', 'ASRS', 'WURS', 'HADS_A', 'HADS_D',
                   'ACC', 'ACC_DAYS', 'HRV', 'HRV_HOURS', 'CPT_II',
                   'ADD', 'BIPOLAR', 'UNIPOLAR', 'ANXIETY', 'CT',
                   'ACC_TIME_SEC', 'HRV_TIME_SEC']

# Ensure columns exist
tabular_features = [col for col in tabular_features if col in patient_info_filtered.columns]
X_tabular = patient_info_filtered[tabular_features].copy()  # Explicit copy

# Scale numeric features
numeric_cols = ['AGE', 'ASRS', 'WURS', 'HADS_A', 'HADS_D',
                'ACC_DAYS', 'HRV_HOURS', 'ACC_TIME_SEC', 'HRV_TIME_SEC']
numeric_cols = [col for col in numeric_cols if col in X_tabular.columns]

if numeric_cols:  # Only scale if we have numeric columns
    scaler = StandardScaler()
    X_tabular.loc[:, numeric_cols] = scaler.fit_transform(X_tabular[numeric_cols])

X_tabular = X_tabular.values

# ==============================================
# STEP 4: Final Validation Before Modeling
# ==============================================
if len(X_activity) == 0 or len(X_tabular) == 0 or len(y) == 0:
    print("\nDATA VALIDATION FAILED:")
    print(f"Activity data samples: {len(X_activity)}")
    print(f"Tabular data samples: {len(X_tabular)}")
    print(f"Target values: {len(y)}")
    raise ValueError("Cannot proceed with empty data arrays")

# ==============================================
# STEP 5: Train-Test Split
# ==============================================
X_activity_train, X_activity_test, X_tabular_train, X_tabular_test, y_train, y_test = train_test_split(
    X_activity, X_tabular, y, test_size=0.2, random_state=42, stratify=y)

# Continue with model building...

# ==============================================
# STEP 6: Build Multi-Input CNN Model
# ==============================================
# Activity data input
activity_input = Input(shape=(1000, 3), name='activity_input')
x = Conv1D(64, kernel_size=10, activation='relu')(activity_input)
x = MaxPooling1D(pool_size=2)(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Conv1D(128, kernel_size=10, activation='relu')(x)
x = MaxPooling1D(pool_size=2)(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Flatten()(x)
activity_output = Dense(64, activation='relu')(x)

# Tabular data input
tabular_input = Input(shape=(X_tabular.shape[1],), name='tabular_input')
tabular_output = Dense(32, activation='relu')(tabular_input)
tabular_output = BatchNormalization()(tabular_output)
tabular_output = Dropout(0.2)(tabular_output)

# Combine both inputs
combined = concatenate([activity_output, tabular_output])
z = Dense(64, activation='relu')(combined)
z = Dropout(0.4)(z)
output = Dense(1, activation='sigmoid')(z)

model = Model(inputs=[activity_input, tabular_input], outputs=output)

model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])

model.summary()

# ==============================================
# STEP 7: Train the Model
# ==============================================
# Calculate class weights
class_weights = {0: 1, 1: len(y_train)/(2*np.bincount(y_train)[1])} if len(np.unique(y_train)) > 1 else {0:1, 1:1}

# Train the model
history = model.fit(
    {'activity_input': X_activity_train, 'tabular_input': X_tabular_train},
    y_train,
    validation_data=(
        {'activity_input': X_activity_test, 'tabular_input': X_tabular_test},
        y_test
    ),
    epochs=100,
    batch_size=32,
    callbacks=[
        EarlyStopping(patience=10, restore_best_weights=True),
        ReduceLROnPlateau(factor=0.1, patience=5)
    ],
    class_weight=class_weights
)

# ==============================================
# STEP 8: Evaluate the Model
# ==============================================
# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()

# Evaluate on test set
test_results = model.evaluate(
    {'activity_input': X_activity_test, 'tabular_input': X_tabular_test},
    y_test
)
print(f"\nTest Loss: {test_results[0]:.4f}")
print(f"Test Accuracy: {test_results[1]:.4f}")
print(f"Test AUC: {test_results[2]:.4f}")

# Make predictions
y_pred = model.predict({'activity_input': X_activity_test, 'tabular_input': X_tabular_test})
y_pred_class = (y_pred > 0.5).astype(int)

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_class))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred_class)
print("Confusion Matrix:")
print(cm)
```

```python
# ==============================================
# STEP 1: Import Required Libraries
# ==============================================
import pandas as pd
import numpy as np
import os
import re
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization, concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ==============================================
# STEP 2: Load and Preprocess Patient Info Data
# ==============================================
def preprocess_patient_data(df):
    # Convert comma decimals to points
    decimal_cols = ['ACC_DAYS', 'HRV_HOURS', 'WURS', 'ASRS', 'MADRS', 'HADS_A', 'HADS_D']
    for col in decimal_cols:
        if col in df.columns:
            df[col] = df[col].astype(str).str.replace(',', '.').astype(float)

    # Convert binary columns
    binary_cols = ['ADHD', 'ADD', 'BIPOLAR', 'UNIPOLAR', 'ANXIETY',
                   'SUBSTANCE', 'OTHER', 'CT', 'MDQ_POS', 'MED']
    for col in binary_cols:
        if col in df.columns:
            df[col] = df[col].fillna(0).astype(int)

    # Fill numeric columns with median
    numeric_cols = ['AGE', 'ACC_DAYS', 'HRV_HOURS', 'WURS', 'ASRS',
                    'MADRS', 'HADS_A', 'HADS_D']
    for col in numeric_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
            df[col] = df[col].fillna(df[col].median())

    # Convert time columns to total seconds with explicit format
    time_cols = ['ACC_TIME', 'HRV_TIME']
    for col in time_cols:
        if col in df.columns:
            df[col] = df[col].astype(str)
            df[col] = df[col].replace('nan', np.nan)
            df[col] = pd.to_datetime(df[col], format='%H:%M:%S', errors='coerce')
            df[col+'_SEC'] = df[col].dt.hour * 3600 + df[col].dt.minute * 60 + df[col].dt.second
            df.drop(col, axis=1, inplace=True)

    # Drop columns with too many missing values
    cols_to_drop = ['MED_Antidepr', 'MED_Moodstab', 'MED_Antipsych',
                   'MED_Anxiety_Benzo', 'MED_Sleep', 'MED_Analgesics_Opioids',
                   'MED_Stimulants']
    for col in cols_to_drop:
        if col in df.columns:
            df.drop(col, axis=1, inplace=True)

    return df

# Load patient info
patient_info = pd.read_csv(
    '/content/drive/MyDrive/hyperaktiv_with_controls/patient_info.csv',
    delimiter=';',
    engine='python'
)
patient_info_clean = preprocess_patient_data(patient_info)

# ==============================================
# STEP 3: Match Patients with Activity Files
# ==============================================
activity_data_path = '/content/drive/MyDrive/hyperaktiv_with_controls/activity_data'

# Get all available activity file IDs
activity_files = os.listdir(activity_data_path)
activity_ids = []

for file in activity_files:
    match = re.search(r'patient_activity_(\d+)\.csv', file)
    if match:
        activity_ids.append(int(match.group(1)))

print(f"Found {len(activity_ids)} activity files")
print(f"First 10 activity file IDs: {activity_ids[:10]}")

# Filter to only patients with activity data
patient_info_filtered = patient_info_clean[patient_info_clean['ID'].isin(activity_ids)].copy()

if len(patient_info_filtered) == 0:
    patient_ids_sample = sorted(patient_info_clean['ID'].unique().tolist()[:20])
    print("\nERROR: No matching patients found between info and activity files")
    print(f"Patient IDs in info file: {patient_ids_sample}...")
    print(f"Activity file IDs: {sorted(activity_ids)[:20]}...")
    raise ValueError("No overlap between patient info and activity files")
else:
    print(f"\nFound {len(patient_info_filtered)} patients with both info and activity data")
    print(f"First 10 matching IDs: {patient_info_filtered['ID'].head(10).tolist()}")

# ==============================================
# STEP 4: Load and Process Activity Data
# ==============================================
def load_activity_data(patient_ids, base_path):
    activity_data = []

    for pid in patient_ids:
        file_path = f"{base_path}/patient_activity_{pid}.csv"
        try:
            df = pd.read_csv(file_path)
            if not df.empty:
                # Select accelerometer columns
                acc_cols = [col for col in df.columns if any(x in col.lower() for x in ['x', 'y', 'z', 'acc'])]
                if len(acc_cols) >= 3:
                    df = df[acc_cols[:3]]

                    # Normalize and fix length
                    scaler = StandardScaler()
                    scaled_data = scaler.fit_transform(df.values)
                    fixed_length = 1000

                    if len(scaled_data) > fixed_length:
                        scaled_data = scaled_data[:fixed_length]
                    else:
                        padding = np.zeros((fixed_length - len(scaled_data), 3))
                        scaled_data = np.vstack([scaled_data, padding])

                    activity_data.append(scaled_data)
                    print(f"✓ Loaded data for patient {pid}")
                else:
                    print(f"× Insufficient columns in {file_path}")
            else:
                print(f"× Empty file: {file_path}")
        except Exception as e:
            print(f"! Error loading {file_path}: {str(e)}")

    if not activity_data:
        raise ValueError("No activity data could be loaded")

    return np.array(activity_data)

X_activity = load_activity_data(patient_info_filtered['ID'], activity_data_path)
y = patient_info_filtered['ADHD'].values

# ==============================================
# STEP 5: Prepare Tabular Features
# ==============================================
# Select features for tabular data
tabular_features = ['SEX', 'AGE', 'ASRS', 'WURS', 'HADS_A', 'HADS_D',
                   'ACC', 'ACC_DAYS', 'HRV', 'HRV_HOURS', 'CPT_II',
                   'ADD', 'BIPOLAR', 'UNIPOLAR', 'ANXIETY', 'CT',
                   'ACC_TIME_SEC', 'HRV_TIME_SEC']

# Ensure we only keep columns that exist
tabular_features = [col for col in tabular_features if col in patient_info_filtered.columns]

X_tabular = patient_info_filtered[tabular_features]

# Scale numeric features
numeric_cols = ['AGE', 'ASRS', 'WURS', 'HADS_A', 'HADS_D',
                'ACC_DAYS', 'HRV_HOURS', 'ACC_TIME_SEC', 'HRV_TIME_SEC']
numeric_cols = [col for col in numeric_cols if col in X_tabular.columns]

scaler = StandardScaler()
X_tabular[numeric_cols] = scaler.fit_transform(X_tabular[numeric_cols])

# Convert to numpy array
X_tabular = X_tabular.values

# ==============================================
# STEP 6: Train-Test Split
# ==============================================
X_activity_train, X_activity_test, X_tabular_train, X_tabular_test, y_train, y_test = train_test_split(
    X_activity, X_tabular, y, test_size=0.2, random_state=42, stratify=y)

# ==============================================
# STEP 7: Build Multi-Input CNN Model
# ==============================================
# Activity data input
activity_input = Input(shape=(1000, 3), name='activity_input')
x = Conv1D(64, kernel_size=10, activation='relu')(activity_input)
x = MaxPooling1D(pool_size=2)(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Conv1D(128, kernel_size=10, activation='relu')(x)
x = MaxPooling1D(pool_size=2)(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Flatten()(x)
activity_output = Dense(64, activation='relu')(x)

# Tabular data input
tabular_input = Input(shape=(X_tabular.shape[1],), name='tabular_input')
tabular_output = Dense(32, activation='relu')(tabular_input)
tabular_output = BatchNormalization()(tabular_output)
tabular_output = Dropout(0.2)(tabular_output)

# Combine both inputs
combined = concatenate([activity_output, tabular_output])
z = Dense(64, activation='relu')(combined)
z = Dropout(0.4)(z)
output = Dense(1, activation='sigmoid')(z)

model = Model(inputs=[activity_input, tabular_input], outputs=output)

model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])

model.summary()

# ==============================================
# STEP 8: Train the Model
# ==============================================
# Calculate class weights
class_weights = {0: 1, 1: len(y_train)/(2*np.bincount(y_train)[1])} if len(np.unique(y_train)) > 1 else {0:1, 1:1}

# Train the model
history = model.fit(
    {'activity_input': X_activity_train, 'tabular_input': X_tabular_train},
    y_train,
    validation_data=(
        {'activity_input': X_activity_test, 'tabular_input': X_tabular_test},
        y_test
    ),
    epochs=100,
    batch_size=32,
    callbacks=[
        EarlyStopping(patience=10, restore_best_weights=True),
        ReduceLROnPlateau(factor=0.1, patience=5)
    ],
    class_weight=class_weights
)

# ==============================================
# STEP 9: Evaluate the Model
# ==============================================
# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()

# Evaluate on test set
test_results = model.evaluate(
    {'activity_input': X_activity_test, 'tabular_input': X_tabular_test},
    y_test
)
print(f"\nTest Loss: {test_results[0]:.4f}")
print(f"Test Accuracy: {test_results[1]:.4f}")
print(f"Test AUC: {test_results[2]:.4f}")

# Make predictions
y_pred = model.predict({'activity_input': X_activity_test, 'tabular_input': X_tabular_test})
y_pred_class = (y_pred > 0.5).astype(int)

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_class))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred_class)
print("Confusion Matrix:")
print(cm)

# ==============================================
# STEP 10: Save the Model (Optional)
# ==============================================
model.save('adhd_cnn_model.h5')
print("Model saved successfully!")
```
